```python
#Inisiasi Cluster Awal
clusters = []
for i in range(len(X)):
    clusters.append([i])
clusters
```

```python
def searchMin(mX):
    titik = []
    minimum = mX[0][1];
#     print("A")
    for i in range(len(mX)):
        for j in range(len(mX[i])):
            if mX[i][j] <= minimum and mX[i][j] != 0:
                minimum = mX[i][j]
                titik = [i,j]
#                 print(titik)
    return titik
# mustMerge = searchMin(mX)
```
